Cell 10 — Convert binary yes/no columns to numeric 1/0 and preview the change.

```python
binary_cols=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']

df[binary_cols] = df[binary_cols].replace({'yes':1,'no':0})

df.head()

```

Cell 11 — One-hot encode the 'furnishingstatus' categorical column and show the result.

```python
df = pd.get_dummies(df, columns=['furnishingstatus'])  
df.head()
```

Cell 12 — Separate target (price) from features (X).

```python
y = df['price']
x = df.drop('price', axis=1)
```

Cell 15 — Define numerical columns to scale, and split data into train/test sets.

```python
# scaler = StandardScaler()

numerical_cols = ['area','bedrooms','bathrooms','stories','parking']

# scaler.fit_transform() => learn data (min , max , mean)

scaler = StandardScaler()
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)

```

Cell 16 — Create a StandardScaler instance (prepared for scaling numeric features).

```python
scaler=StandardScaler()
```

Cell 17 — Fit scaler on training numerical columns and transform both train and test numeric columns.

```python
x_train[numerical_cols]=scaler.fit_transform(x_train[numerical_cols])

x_test[numerical_cols]=scaler.transform(x_test[numerical_cols])
```

Cell 20 — Initialize the LinearRegression model instance.

```python
model = LinearRegression()
```

Cell 21 — Fit the LinearRegression model on training data.

```python
model.fit(x_train,y_train)
```

Cell 22 — Predict on test set and compute Mean Absolute Error as evaluation metric.

```python
y_pred = model.predict(x_test)

print(f"Mean Absoulte Error: {mean_absolute_error(y_test,y_pred):.2f}")
```
