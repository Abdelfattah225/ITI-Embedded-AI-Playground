```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error


```

```python
np.random.seed(42)
x = np.random.rand(100,1) * 10
y = 2.5* x**2 + .5 * x + np.random.randn(100,1)*50 #quadretic func
plt.figure(figsize=(10, 5))
plt.scatter(x, y, alpha=0.7)
plt.title("Data with HIGH Noise (×50)")
plt.xlabel("x")
plt.ylabel("y")
```

```python
x_train , x_test , y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

```

```python

# ═══════════════════════════════════════════
# UNDERFITTING: Use degree=1 (linear) for quadratic data!
# ═══════════════════════════════════════════

# to simple 
poly_under = PolynomialFeatures(degree=1)
x_train_under = poly_under.fit_transform(x_train)
x_test_under = poly_under.transform(x_test)

model = LinearRegression()

model.fit(x_train_under,y_train)
```

```python
y_train_pred_under = model.predict(x_train_under)
y_test_pred_under = model.predict(x_test_under)

```

```python
print("=" * 50)
print(" UNDERFITTING MODEL (degree=1) ")
print("=" * 50)
print(f"Train R^2: {r2_score(y_train,y_train_pred_under)}")
print(f"Test R^2: {r2_score(y_test,y_test_pred_under)}")
print(f"Train MSE: {mean_squared_error(y_train, y_train_pred_under):.2f}")
print(f"Test MSE:  {mean_squared_error(y_test, y_test_pred_under):.2f}")
```
