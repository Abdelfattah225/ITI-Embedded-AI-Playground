```python
from tensorflow.keras.models import Sequential  # Capital 'S'
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.regularizers import l2  # Fixed spelling

# If need to modify Adam parameters
# if need to modify adam para
#optimizer = Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, amsgrad=False)

from tensorflow.keras.optimizers import Adam
# optimizer = Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, amsgrad=False)


# Fix Option 1: Convert to one-hot
from tensorflow.keras.utils import to_categorical

from sklearn.datasets import make_classification  # sklearn, not tensorflow
from sklearn.model_selection import train_test_split

```

```python
model = Sequential()
# input layer with L2 regularization
model.add(
    Dense(64, activation='relu', input_shape=(100,), kernel_regularizer=l2(0.01))
)
# hidden layer with dropout and L2 regularization
model.add(Dense(32, activation='relu', kernel_regularizer=l2(0.01)))
# output layer
model.add(Dense(10, activation='softmax'))

```

```python
# build layer
model.compile(
    optimizer=Adam(learning_rate=0.001),
       loss='sparse_categorical_crossentropy',  # Works with integer labels,
    metrics=['accuracy']
)
```

```python
model = Sequential()
model.add(Dense(128, input_dim=784, activation='relu', kernel_regularizer=l2(0.001)))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.001)))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# generate synthetic data
x,y=make_classification(n_samples=1000, n_features=784, n_classes=10, n_informative=50, random_state=42)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

y_train_encoded = to_categorical(y_train, num_classes=10)
y_test_encoded = to_categorical(y_test, num_classes=10)
```
