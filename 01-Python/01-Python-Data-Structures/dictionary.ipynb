```python
# nested dict

stuents = {
    "student1":{
        "name":"abdo",
        "age" : "25",
        "grades":[400,254,45]
    },
    "student2":{
        "name":"hussien",
        "age" :"140",
        "grades":[6,4,1]
    }
}

print(stuents["student1"]["grades"][0])
```

```python
# dict in range
squares = {x:x**2 for x in range(6)}
squares
```

```python
squares = {x:x**2 for x in range(10) if x%2==0}
squares
```
