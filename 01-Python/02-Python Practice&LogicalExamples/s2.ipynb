```python
frist = 0    # 0th Fibonacci
second = 1   # 1st Fibonacci
next = 0     # Initialize to print 0th first
counter = 0

print(f"fibonacci {counter} is {frist}")  # Print 0th explicitly first

while counter < 10:  # Changed <=10 to <10 to print exactly 11 numbers (0-10)
    next = frist + second
    frist = second
    second = next
    counter += 1
    print(f"fibonacci {counter} is {next}")
```
