✏️ Task 7:
Create a function that categorizes Age:

Age < 20: "Young"
Age 20-30: "Adult"
Age > 30: "Senior"
Apply it to create a new column 'Age_Category'

```python
def categorizes_Age(x):
    if x<20:
        return "Young"
    elif x <= 30:
        return "Adult"
    else:
        return "Senior"
    
dataFrame['Age_Category'] = dataFrame['Age'].apply(categorizes_Age)

dataFrame
```
