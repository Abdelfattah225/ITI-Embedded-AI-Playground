```python
import pandas as pd

import numpy as np
```

```python
employees = {
    "Name": ["Ahmed", "Sara", None, "Mohamed", "Fatima", "Ali"],
    "Department": ["IT", "HR", "IT", "Finance", None, "IT"],
    "Salary": [5000, 4500, 6000, None, 4000, 5500],
    "Age": [28, 35, 24, 45, 30, None],
    "Joining_Date": ["2020-03-15", "2019-07-20", "2022-01-10", 
                     "2018-05-25", "2021-08-30", "2020-11-15"]
}
df = pd.DataFrame(employees)
```

```python
df['Department'].fillna('Unknown',inplace=True)
```

```python
df
```

```python
df['Salary'].fillna(df['Salary'].mean(), inplace=True)

df
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df
```

```python
df[['Name','Salary']]
```

```python
df[df['Salary']>5000]
```

```python
df[(df['Department']=='IT') & (df['Age'] < 30)]
```

```python
df[(df['Department']=='HR') | (df['Department'] == 'Finance')]
```

```python
df['Bonus'] = df['Salary'] * .2 
df['Tax']  = df['Salary'] * .1
df['Net_Salary'] = df['Salary'] + df['Bonus'] - df['Tax']
```

```python
df
```

```python
def Salary_Category(x):
    if x < 4500:
        return "Low"
    elif x <= 5500:
        return "Medium"
    else:
        return "High"
    

df['Salary_Category'] = df['Salary'].apply(Salary_Category)


df
```
