```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp = [25,25.5,26,26.5,27,27.5,30]
time= [1,2,3,4,5,6,7]

plt.plot(time, temp)
plt.title("Temp change over time ")
plt.xlabel("Time(seconds)")

plt.ylabel("temp (c)")
plt.show()
```

```python
soil_moist = [10,20,30,40,50,60,70]
temp = [25,25.5,26,26.5,27,27.5,30]
plt.scatter(soil_moist,temp)
plt.title("Relation between soil and temperture")
plt.xlabel("Soil moisture(%)")
plt.ylabel("Temperture(c)")

plt.show()
```

```python
#bar charts
categories=['A', 'B', 'C' , 'D']
values = [4,5,7,9]


plt.bar(categories,values,color='green')
plt.title("Freq of categories")
plt.xlabel("categories")
plt.ylabel("Frequency")
plt.show()
```

```python
#histogram


data = [1,2,2,3,3,3,4,4,4,4]
plt.hist(data,bins=4,color='blue')
plt.title("Simple histogram")
plt.ylabel("Frequency")
plt.xlabel("Bins")
plt.show()
```

```python
#sub plot determined tall of plot  , postion , widith 
plt.subplot(1,2,1)
plt.plot([1,2,3,4,5],[1,5,8,10,6])
plt.title("Plot1")

plt.subplot(1,2,2)
plt.plot([2,4,6,8,10],[4,6,7,8,10])
plt.title("Plot2")

plt.show()
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.lineplot(x='total_bill', y='tip', data=df)
plt.title("Line Plot")
plt.xlabel("total bill ")
plt.ylabel("Tips")
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=df)


```

```python
sns.countplot(x='sex',data=df)

plt.title("count plot of sex")

plt.show()
```

```python
sns.boxenplot(x='day',y='total_bill',data=df)

plt.title("box plot")#

plt.show()
```

```python
numerical_cols = df.select_dtypes(['int64','float64'])


numerical_cols.head()
```

```python

#correlation matrix

numerical_cols.corr()
```

```python
sns.heatmap(numerical_cols.corr(),annot=True, cmap='Blues')
```
