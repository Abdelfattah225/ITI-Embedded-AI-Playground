```python
Days = [1, 2, 3, 4, 5]
Sales = [100, 150, 130, 180, 200]

plt.plot(Sales,Days)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Linear plot")
plt.show()
```

```python
Days = [1, 2, 3, 4, 5]
Sales = [100, 150, 130, 180, 200]

plt.scatter(Sales,Days)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Linear plot")
plt.show()
```

 Task 3:
Create a 2x2 subplot layout with:

Position 1: Line plot
Position 2: Scatter plot
Position 3: Bar chart
Position 4: Histogram

```python
fig, axs = plt.subplots(2,2)
axs[0,0] = plt.plot(([1,2,3,4],[2,4,6,8]))
axs[0,0].set_title('Line Plot')

axs[0,1] = plt.scatter([1,2,3,4],[2,4,6,8])
axs[0,1].set_title("scatter Plot")


ruits = ['Apple', 'Banana', 'Orange']
Counts = [30, 45, 25]
axs[1,0] = plt.bar(ruits,Counts)
axs[1,0].set_title("bar Plot")


data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
axs[1,1] = plt.hist(data, bins=4)
axs[1,1].set_title("histgram Plot")
```
